```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.datasets import make_classification
```

```python
def logistic(w, b, x):
    return 1. / (1. + np.exp(-(x.dot(w)) + b))
```

```python
x = np.arange(-10, 10, 0.01)
```

```python
data = np.arange(-10, 10, 0.5)
labels = np.random.binomial(1, logistic(-0.5, 2, data))
plt.title('Labeled Dataset with 1-D Features')
plt.ylabel('y')
plt.xlabel('x')
plt.scatter(data, labels, c=labels)
plt.savefig('figures/figure_1.png')
```

```python
plt.title('Logistic Curve')
plt.plot(x, logistic(1.0, 0, x))

plt.scatter(data, labels, c=labels)

plt.xlabel('$x$')
plt.ylabel('$\sigma(x)$')
plt.savefig('figures/figure_2.png')
```

```python
plt.title('Logistic Curves with Varying Positive Weights')
plt.plot(x, logistic(1.0, 0, x), label='w = 1.0')
plt.plot(x, logistic(0.5, 0, x), label='w = 0.5')
plt.plot(x, logistic(2.0, 0, x), label='w = 2.0')

plt.scatter(data, labels, c=labels)

plt.xlabel('$x$')
plt.ylabel('$\sigma_w(x)$')
plt.legend()
plt.savefig('figures/figure_3.png')
```

```python
plt.title('Logistic Curves with Varying Negative Weights')
plt.plot(x, logistic(-1.0, 0, x), label='w = -1.0')
plt.plot(x, logistic(-0.5, 0, x), label='w = -0.5')
plt.plot(x, logistic(-2.0, 0, x), label='w = -2.0')

plt.scatter(data, labels, c=labels)

plt.xlabel('$x$')
plt.ylabel('$\sigma_w(x)$')
plt.legend()
plt.savefig('figures/figure_4.png')
```

```python
plt.title('Logistic Curves with Varying Biases')
plt.plot(x, logistic(-1.0, 0.0, x), label='b = 0.0')
plt.plot(x, logistic(-1.0, 2.0, x), label='b = 2.0')
plt.plot(x, logistic(-1.0, -2.0, x), label='b = -2.0')

plt.scatter(data, labels, c=labels)

plt.xlabel('$x$')
plt.ylabel('$\sigma_b(x)$')
plt.legend()
plt.savefig('figures/figure_5.png')
```

```python
plt.title('Correct Logistic Curve')
plt.plot(x, logistic(-0.5, 2.0, x), label='correct')

plt.scatter(data, labels, c=labels)

plt.xlabel('$x$')
plt.ylabel(r'$\sigma_\theta(x)$')
plt.legend()
plt.savefig('figures/figure_5a.png')
```

```python
plt.title('Convex Loss Surface')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$L(y, \sigma_\theta(x))$')
plt.plot(x, x ** 2)
plt.savefig('figures/figure_10.png')
```

```python
plt.title('Non-convex Loss Surface')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$L(y, \sigma_\theta(x))$')
plt.plot(x, 0.4*np.sin(x)+np.sin(1/50 * x**2)+1/30*x)
plt.savefig('figures/figure_11.png')
```

```python
def cross_entropy(y, y_hat):
    return - y * np.log(y_hat)

ys = np.arange(0.001, 1., 0.001)
plt.title('Cross-Entropy Loss for a Single Instance (y = 1)')
plt.xlabel('$\hat{y}$')
plt.ylabel('$L(y, \hat{y})$')
plt.plot(ys, cross_entropy(1.0, ys))
plt.savefig('figures/figure_9.png')
```

```python
def mse(y, y_hat):
    return (y - y_hat) ** 2

ys = np.arange(0.001, 1., 0.001)
plt.title('Mean Squared Error (MSE) Loss for a Single Instance (y = 1)')
plt.xlabel('$\hat{y}$')
plt.ylabel('$L(y, \hat{y})$')
plt.plot(ys, mse(1.0, ys))
plt.savefig('figures/figure_8.png')
```

```python
cluster1 = 1.2 * np.random.randn(100, 2) + 0.5
cluster2 = 1.2 * np.random.randn(100, 2) -3
```

```python
plt.title('2-Dimensional Labeled Data')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.plot(cluster1[:, 0], cluster1[:, 1], 'o', c='green', label='y = 0')
plt.plot(cluster2[:, 0], cluster2[:, 1], 'P', c='purple', label='y = 1')
plt.legend()
plt.savefig('figures/figure_6.png')
```

```python
plt.title('2-Dimensional Labeled Data, Separated')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.plot(cluster1[:, 0], cluster1[:, 1], 'o', c='green')
plt.plot(cluster2[:, 0], cluster2[:, 1], 'P', c='purple')
plt.plot([3, -6], [-5, 3.4], 'k-', c='black')
plt.savefig('figures/figure_7.png')
```

```python
plt.title('XOR')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.plot([0, 1], [0, 1], 'X', c='green', label='y = 0')
plt.plot([1, 0], [0, 1], 'o', c='purple', label='y = 1')
plt.legend()
plt.savefig('figures/figure_14.png')
```
